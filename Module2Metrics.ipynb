```python
import json
import networkx as nx

g=nx.Graph()

file_path = "eightyLVL_tag.json"
```

```python
with open (file_path, "r") as in_file:
  all_article_tags = json.load(in_file)
for url, tags in all_article_tags.items():
  for tag in tags:
    if tag not in g:
      g.add_node(tag)

  i=0
  for left_tag in tags:
    for right_tag in tags[i+1:]:
      current_weight = g[left_tag][right_tag]['weight'] if g.has_edge(left_tag, right_tag) else 0
      g.add_edge(left_tag, right_tag, weight=current_weight + 1)
    i += 1




print('nodes (tags):', len(g.nodes))
print('edge (occurence):', len(g.edges))
```

```python
nx.write_graphml(g,"tags.graphml")
print("GraphML exported to: tags.graphml")
```
