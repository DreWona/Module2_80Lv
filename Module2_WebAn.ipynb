```python
import tldextract # Useful for extracting domains from URLs
url = "https://80.lv/articles"
def getDomain(url):
    """
    Use the tldextract package to get the domain name for a given URL
    """

    extracted_domain = tldextract.extract(url)
    return extracted_domain.fqdn
```

```python
from bs4 import BeautifulSoup
import re
```

```python
import requests
```

```python
r = requests.get('https://80.lv/articles') # Run an HTTP GET request

# Check for success.
#. If successful, we'll get a 200 status code
#. HTTP has many such codes, which you can reference online (e.g., 404, 500)
print(r.status_code)
```

```python
# Use the Encoding supplied by the server response
#. to convert the content to a string, which should
#. produce the HTML we want to analyze
article_html = r.content.decode(r.encoding)
print("Length:", len(article_html))

print(article_html[:256])
```

# New Section

```python
# MAKE PRETTY
soup = BeautifulSoup(article_html, 'html.parser')

print(soup.prettify())
```

```python
# # BS lets us find all HTML tags of a particular type
# #. For links, we care about <a href...> style
# #. (though we could also look for img links or anything that
# #. links to content)

# # Here, we use .findAll() to find all <a href...> tags

# edges = []
# for link in soup.find_all('a'):
#     if link.has_attr('href'):
#         link_url = link.get('href')

#         print("\t", "Raw Link:", link_url)

#         if link_url.startswith("http"):
#             # process this link
#             print("\t", "Off-site Link:", getDomain(link_url))
#         elif link_url.startswith("//"): # Special case where link excludes the HTTP
#             # process this link
#             print("\t", "Off-site Link:", getDomain("http:" + link_url))
#         else: # Not a web link to another article?
#             pass
```

```python
edges = []  # will store (source, target) tuples

for link in soup.find_all('a'):
    if link.has_attr('href'):
        link_url = link.get('href')

        if link_url.startswith("http") and '80.lv/articles/' in link_url:
            edges.append(("https://80.lv/articles", link_url))
            print("Article Link Found:", link_url)

        elif link_url.startswith("//") and '80.lv/articles/' in link_url:
            full = "http:" + link_url
            edges.append(("https://80.lv/articles", full))
            print("Article Link Found:", full)

print(f"\nTotal article links found: {len(edges)}")
```

```python
# Step 1: Get all article URLs from the main page
article_urls = []

for link in soup.find_all('a'):
    if link.has_attr('href'):
        link_url = link.get('href')
        if link_url.startswith('/articles/') and len(link_url) > len('/articles/'):
            full_url = 'https://80.lv' + link_url
            article_urls.append(full_url)

article_urls = list(set(article_urls))
print(f"Found {len(article_urls)} articles on main page")
for a in article_urls:
    print(" ", a)
```

```python
import time
```

```python
# Step 2: Visit each article and collect its tags
article_tags = {}

for url in article_urls:
    try:
        r_article = requests.get(url, timeout=5)
        soup_article = BeautifulSoup(r_article.content.decode(r_article.encoding), 'html.parser')

        tags = []
        for link in soup_article.find_all('a'):
            if link.has_attr('href'):
                link_text = link.text.strip()
                link_url = link.get('href')
                if link_text.startswith('#') and link_url.startswith('/articles/'):
                    tags.append(link_text)

        article_tags[url] = tags
        print(f"Article: {url.split('/articles/')[1][:40]} | Tags: {tags}")
        time.sleep(0.5)

    except Exception as e:
        print(f"Failed: {url} â€” {e}")

print(f"\nDone. Collected tags for {len(article_tags)} articles")
```

```python
# Collect all unique tags from our already-crawled articles
all_tags = set()

for url, tags in article_tags.items():
    for tag in tags:
        all_tags.add(tag)

print(f"Found {len(all_tags)} unique tags:")
for tag in sorted(all_tags):
    print(" ", tag)
```

```python
# Convert tags to their URLs
tag_urls = {}

for tag in all_tags:
    # Remove # and convert spaces to hyphens, lowercase
    tag_slug = tag.replace('#', '').strip().lower().replace(' ', '-')
    tag_url = f'https://80.lv/articles/{tag_slug}'
    tag_urls[tag] = tag_url

print("Tag URLs:")
for tag, url in sorted(tag_urls.items()):
    print(f"  {tag} -> {url}")
```

```python
# Test one tag page
r_tag = requests.get('https://80.lv/articles/blender')
soup_tag = BeautifulSoup(r_tag.content.decode(r_tag.encoding), 'html.parser')

# Find articles on this tag page
test_articles = []
for link in soup_tag.find_all('a'):
    if link.has_attr('href'):
        link_url = link.get('href')
        link_text = link.text.strip()
        if link_url.startswith('/articles/') and not link_url.startswith('#') and link_text and not link_text.startswith('#'):
            test_articles.append(link_url)

test_articles = list(set(test_articles))
print(f"Found {len(test_articles)} articles on Blender tag page:")
for a in test_articles[:10]:
    print(" ", a)
```

```python
# Crawl all tag pages and collect articles
import time

tag_articles = {}  # { tag: [list of article URLs] }

for tag, url in tag_urls.items():
    try:
        r_tag = requests.get(url, timeout=5)
        soup_tag = BeautifulSoup(r_tag.content.decode(r_tag.encoding), 'html.parser')

        articles = []
        for link in soup_tag.find_all('a'):
            if link.has_attr('href'):
                link_url = link.get('href')
                link_text = link.text.strip()

                # Filter out tag pages and non-articles
                if (link_url.startswith('/articles/')
                    and link_text
                    and not link_text.startswith('#')
                    and link_url != f'/articles/{tag.replace("#","").strip().lower().replace(" ","-")}'
                    and link_url != '/articles/'):

                    full_url = 'https://80.lv' + link_url
                    articles.append(full_url)

        articles = list(set(articles))
        tag_articles[tag] = articles
        print(f"{tag} -> {len(articles)} articles")
        time.sleep(0.5)

    except Exception as e:
        print(f"Failed: {tag} â€” {e}")

print(f"\nDone. Crawled {len(tag_articles)} tag pages")
```

```python
# Visit each unique article and collect its tags
all_article_tags = {}  # { article_url: [list of tags] }

# First get all unique article URLs across all tags
all_article_urls = set()
for articles in tag_articles.values():
    for url in articles:
        all_article_urls.add(url)

print(f"Total unique articles to crawl: {len(all_article_urls)}")

# Now visit each article and get its tags
for i, url in enumerate(all_article_urls):
    try:
        r_article = requests.get(url, timeout=5)
        soup_article = BeautifulSoup(r_article.content.decode(r_article.encoding), 'html.parser')

        tags = []
        for link in soup_article.find_all('a'):
            if link.has_attr('href'):
                link_text = link.text.strip()
                link_url = link.get('href')
                if link_text.startswith('#') and link_url.startswith('/articles/'):
                    tags.append(link_text)

        all_article_tags[url] = tags
        print(f"[{i+1}/{len(all_article_urls)}] {url.split('/articles/')[1][:40]} | Tags: {tags}")
        time.sleep(0.5)

    except Exception as e:
        print(f"Failed: {url} â€” {e}")

print(f"\nDone. Collected tags for {len(all_article_tags)} articles")
```
